Per a respondre a la primera pregunta hem decidit agrupar els accidents per any i dia, per poder comparar quants accidents va haver-hi cada dia en els diferents anys.

```python
count_by_day = data.groupby(['DAY','YEAR']).size().reset_index(name='TOTAL ACCIDENTS')
```

```python
count_by_day
```

```python
alt.Chart(count_by_day).mark_line().encode(
    x='DAY:N',
    y='TOTAL ACCIDENTS',
    color='YEAR:N'
).properties(title = "Comparison of total accidents per day in April between years")
```

Aquesta informació ja contesta parcialment la primera pregunta i la segona, però, vistes les similituds de les dues primeres preguntes, el nostre objectiu es conseguir contestar-les amb una única DataVis.
A continuació agruparem els ferits i morts per dia

```python
injured_by_day = data.groupby(['DAY','YEAR'])['NUMBER OF PERSONS INJURED'].agg('sum').reset_index(name='TOTAL INJURED')
```

```python
injured_by_day
```

```python
alt.Chart(injured_by_day).mark_line().encode(
    x='DAY:N',
    y='TOTAL INJURED:Q',
    color='YEAR:N'
).properties(title = "Comparison of total injured per day in April between years")
```

```python
killed_by_day = data.groupby(['DAY','YEAR'])['NUMBER OF PERSONS KILLED'].agg('sum').reset_index(name='TOTAL KILLED')
```

```python
killed_by_day
```

```python
alt.Chart(killed_by_day).mark_line().encode(
    x='DAY:N',
    y='TOTAL KILLED:N',
    color='YEAR:N'
).properties(title = "Comparison of total killed per day in April between years")
```
