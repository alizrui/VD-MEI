```python
filtered_db.loc[filtered_db['CONTRIBUTING FACTOR VEHICLE 1'] == 'Following Too Closely', 'GENERAL FACTORS 1'] = 'DANGEROUS DRIVING'
```

```python
factor1 = ['Failure to Keep Right', 'Driverless/Runaway Vehicle',  'Following Too Closely', 'Passing Too Closely', 'Failure to Yield Right-of-Way', 'Turning Improperly', 'Unsafe Speed', 'Passing or Lane Usage Improper', 'Traffic Control Disregarded', 'Aggressive Driving/Road Rage', 'Unsafe Lane Changing']
factor2 = ['Other Electronic Device', 'Eating or Drinking', 'Texting', 'Cell Phone (hand-Held)', 'Using On Board Navigation Device', 'Cell Phone (hands-free)', 'Passenger Distraction','Driver Inattention/Distraction', 'Outside Car Distraction' ]
factor3 = ['Shoulders Defective/Improper', 'Prescription Medication', 'Physical Disability', 'Fatigued/Drowsy', 'Lost Consciousness', 'Illnes', 'Fell Asleep']
factor4 = ['Drugs (illegal)', 'Alcohol Involvement']
factor5 = ['Driver Inexperience']
factor6 = ['Vehicle Vandalism', 'Headlights Defective', 'Windshield Inadequate', 'Tow Hitch Defective', 'Accelerator Defective', 'Tire Failure/Inadequate', 'Brakes Defective', 'Steering Failure', 'Oversized Vehicle']
factor7 = ['Traffic Control Device Improper/Non-Working', 'Lane Marking Improper/Inadequate','Obstruction/Debris', 'Pavement Slippery', 'Pavement Defective']
factor8 = ['Tinted Windows', 'Glare', 'View Obstructed/Limited', 'Other Lighting Defects']
factor9 = ['Other Vehicular', 'Reaction to Uninvolved Vehicle']
factor10 = ['Pedestrian/Bicyclist/Other Pedestrian Error/Confusion', 'Animals Action']
factor11 = ['Unspecified']

def general_factors(x):
  res = ''
  if pd.isna(x):
    res = x
  elif x in factor1:
    res = 'DANGEROUS DRIVING'
  elif x in factor2:
    res = 'DRIVER DISTRACTION'
  elif x in factor3:
    res = 'DRIVER MEDICAL ISSUE'
  elif x in factor4:
    res = 'DRIVER ON ALCOHOL OR DRUGS'
  elif x in factor5:
    res = 'DRIVER INEXPERIENCE'
  elif x in factor6:
    res = 'CAR MECHANICAL PROBLEMS'
  elif x in factor7:
    res ='STREET AND ROAD PROBLEMS'
  elif x in factor8:
    res = "VISIBILITY PROBLEMS"
  elif x in factor9:
    res = 'OTHER VEHICLE'
  elif x in factor10:
    res = 'EXTERNAL HUMAN OR ANIMAL FACTOR'
  else:
    res = 'UNSPECIFIED'
  return res

```

```python
filtered_db['GENERAL FACTORS 1'] = filtered_db['CONTRIBUTING FACTOR VEHICLE 1'].apply(general_factors)
```
